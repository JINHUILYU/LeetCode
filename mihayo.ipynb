```python
s = input().strip()
k = int(input().strip())

# 对于给定的字符串 s，将其进行k轮改造，第i轮改造的步骤如下:
# - 对于每一个大写字母，将其修改，为其字母表中前面的一位字母，特别的，'A’前面一位为'Z’;
# - 对于每一个小写字母，将其修改为其字母表中后面的第i位字母(i从1开始)，特别的，'z’后面一位为'a’;
# 输出k轮改造后的字符串。

for i in range(1, k + 1):
    transformed = []
    for char in s:
        if 'A' <= char <= 'Z':  # Uppercase letters
            if char == 'A':
                transformed.append('Z')
            else:
                transformed.append(chr(ord(char) - 1))
        elif 'a' <= char <= 'z':  # Lowercase letters
            offset = (ord(char) - ord('a') + i) % 26
            transformed.append(chr(ord('a') + offset))
        else:  # Non-alphabetic characters remain unchanged
            transformed.append(char)
    s = ''.join(transformed)

print(s)
```

```python
# 有两个字符串，分别是s和t。
# 由于数据安全的原因，字符串t不允许作为s的子串存在。
# 为了满足数据规范，米小游需要按照一定的顺序删除字符串s中的一些字符，确保s中不再包含t。
# 注意删除字符si后，不会自动将前后字符串合并，你可以认为使用一个空白字符代替 si。
# 米小游想知道，在保证s不包含t的前提下，她最少需要删除多少个字符?
# 删除顺序可以描述为一个排列，ai 表示删除 sai 字符。
# 示例
# 5 2
# ababa
# ab
# 3 2 5 4 1
# 输出
# 2
def min_deletions_to_remove_substring(s: str, t: str, order: list[int]) -> int:
    n, m = len(s), len(t)
    # delete_time[i] = 在第几次操作中删除 s[i]；没删到则为 +inf
    delete_time = [float('inf')] * n
    for time, idx in enumerate(order):
        delete_time[idx - 1] = time

    # 检查：前 k 次删除后，s 中是否仍包含 t 作为连续子串？
    def check(k: int) -> bool:
        # 对每个可能的起始位置 i，尝试匹配 t
        for i in range(n - m + 1):
            match = True
            for j in range(m):
                # 如果第 i+j 位已被删除（delete_time < k），或字符不等，则不匹配
                if delete_time[i + j] < k or s[i + j] != t[j]:
                    match = False
                    break
            if match:
                return True
        return False

    # 二分：找最小的 k，使得 check(k) == False
    low, high, ans = 0, n, n
    while low <= high:
        mid = (low + high) // 2
        if check(mid):
            # 仍能匹配，则需要更多删除
            low = mid + 1
        else:
            # 已不含 t，尝试更少删除
            ans = mid
            high = mid - 1

    return ans




# 示例输入
s = "ababa"
t = "ab"
order = [3, 2, 5, 4, 1]  # 删除顺序

print(min_deletions_to_remove_substring(s, t, order))  # 输出应为 2
```

```python
# 示例输入
s = "ababa"
t = "ab"
order = [3, 2, 5, 4, 1]  # 删除顺序
cnt = 0
if t not in s:
    print(cnt)
else:
    for i in order:
        s = s[:i - 1] + ' ' + s[i:]
        cnt += 1
        print(s)
        if t not in s:
            print(cnt)
            break
    else:
        # 如果删除完了都没删掉，说明删不掉
        print(-1)
```

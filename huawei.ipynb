````python
import sys
from collections import deque

def main(relations):
    # n = int(sys.stdin.readline())
    mp = dict()  # 邻接表
    d = dict()   # 入度统计
    cnts = dict()  # 迭代次数
    st = set()  # 所有版本集合

    for relation in relations:
        a, b = relation.strip().split()
        if b != "NA":
            if b not in mp:
                mp[b] = set()
            mp[b].add(a)
            d[a] = d.get(a, 0) + 1
        st.add(a)
        st.add(b)

    q = deque()
    for t in st:
        if d.get(t, 0) == 0 and t != "NA":
            q.append(t)
            cnts[t] = 0

    while q:
        t = q.popleft()
        if t in mp:
            for next_node in mp[t]:
                cnts[next_node] = cnts[t] + 1
                q.append(next_node)

    max_cnt = 0
    for cnt in cnts.values():
        if cnt > max_cnt:
            max_cnt = cnt

    res = []
    for key, value in cnts.items():
        if value == max_cnt:
            res.append(key)

    res.sort()
    return ' '.join(res)

if __name__ == "__main__":
    # 测试用例列表
    test_cases = [
        ["CN001 BF001", "BF001 AZ001", "AZ001 NA", "BF010 AZ001", "AW001 NA", "BF011 AZ001"], # CN001
        ["BF0001 AZ0001", "AZ0001 NA", "BF00011 AZ0001"], # BF0001 BF00011
        ["A1 NA", "A2 NA", "B1 A1", "B2 A2", "C1 B1", "C2 B2"], # C1 C2
        ["ONLY1 NA"], # ONLY1
    ]

    # 自动测试并输出结果
    for idx, relation in enumerate(test_cases):
        result = main(relation)
        print(f"Test Case {idx + 1} Output: {result}")
````

### 题目描述

### 输出描述

#### 样例解释

```cpp
#include <bits/stdc++.h>
using namespace std;
const int INF = 1e9;

int main() {
    int N;
    if (!(cin >> N)) return 0;

    char cs, ct;
    cin >> cs >> ct;

    // 1. 读入边，建立映射和邻接表
    map<char,int> mp;
    vector<char> rev; // 反向索引：编号->字符
    vector<tuple<char,char,int>> edges;
    while (true) {
        string a, b;
        int w;
        cin >> a;
        if (a == "0000") break;
        cin >> b >> w;
        edges.emplace_back(a[0], b[0], w);
        // 建映射
        for (char c : {a[0], b[0]}) {
            if (!mp.count(c)) {
                int id = mp.size();
                mp[c] = id;
                rev.push_back(c);
            }
        }
    }

    int V = mp.size();
    vector<vector<pair<int,int>>> adj(V);
    for (auto &[u, v, w] : edges) {
        int ui = mp[u], vi = mp[v];
        adj[ui].emplace_back(vi, w);
        adj[vi].emplace_back(ui, w);
    }

    // 2. Dijkstra 初始化
    int s = mp[cs], t = mp[ct];
    vector<int> dist(V, INF), parent(V, -1);
    dist[s] = 0;
    // min-heap: (距离, 顶点)
    priority_queue<pair<int,int>, vector<pair<int,int>>, greater<>> pq;
    pq.emplace(0, s);

    // 3. 松弛过程
    while (!pq.empty()) {
        auto [d, u] = pq.top(); pq.pop();
        if (d > dist[u]) continue;
        for (auto &[v, w] : adj[u]) {
            if (dist[v] > d + w) {
                dist[v] = d + w;
                parent[v] = u;
                pq.emplace(dist[v], v);
            }
        }
    }

    // 4. 路径还原
    vector<char> path;
    for (int cur = t; cur != -1; cur = parent[cur]) {
        path.push_back(rev[cur]);
    }
    reverse(path.begin(), path.end());

    // 5. 输出
    for (int i = 0; i < (int)path.size(); i++) {
        if (i) cout << ' ';
        cout << path[i];
    }
    cout << '\n';
    return 0;
}
```

````python
import heapq
from typing import Dict, List, Tuple

def fun(input_str: str) -> str:
    """
    读取输入，构建无向带权图，使用 Dijkstra 算法计算最短路径，
    并返回从出发站到目标站的路径，节点间以空格分隔。
    """
    lines = input_str.strip().splitlines()
    it = iter(lines)
    n = int(next(it))  # 站点总数（不直接使用，但可用于校验）
    cs, ct = next(it).split()  # 出发站点、到达站点

    # 建图：用邻接表，节点用字符直接做键
    adj: Dict[str, List[Tuple[str, int]]] = {}
    for line in it:
        if line == '0000':
            break
        u, v, w = line.split()
        w = int(w)
        for x in (u, v):
            adj.setdefault(x, [])
        adj[u].append((v, w))
        adj[v].append((u, w))

    # Dijkstra 最短路径
    dist: Dict[str, int] = {node: float('inf') for node in adj}
    parent: Dict[str, str] = {}
    dist[cs] = 0
    pq: List[Tuple[int, str]] = [(0, cs)]

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        if u == ct:
            break
        for v, w in adj[u]:
            nd = d + w
            if nd < dist[v]:
                dist[v] = nd
                parent[v] = u
                heapq.heappush(pq, (nd, v))

    # 还原路径
    path: List[str] = []
    cur = ct
    while True:
        path.append(cur)
        if cur == cs:
            break
        cur = parent[cur]
    path.reverse()
    return ' '.join(path)


if __name__ == '__main__':
    # 自动化测试
    test_cases = [
        (
            '''12
a e
a b 2
b c 2
c d 2
d e 2
f b 3
b g 3
g h 2
h i 3
i h 2
h e 3
e k 2
k l 4
0000''',
            'a b c d e'
        ),
        (
            '''12
f k
a b 2
b c 2
c d 2
d e 2
f b 3
b g 3
g h 2
h i 3
i h 2
h e 3
e k 2
k l 4
0000''',
            'f b c d e k'
        )
    ]

    for idx, (inp, expected) in enumerate(test_cases, 1):
        result = fun(inp)
        assert result == expected, (
            f"Test case {idx} failed:\n"
            f"  Expected: {expected}\n"
            f"  Got:      {result}"
        )
        print(f"Test case {idx} passed: {result}")
````

# 安排座位 中等

一列具有 m 个座位的火车，从起点到终点共停靠 n 个站点，站点编号从 0 到 n−1。发车前有 x 名乘客预定了座位，因为预定数量可能超出座位数，为了保证效率最大化，请计算如何分配才能使座位利用率最大，并输出最大的座位利用率。

**座位利用率定义**：每个座位被使用的站数。例如有两个座位，第一个座位从第 0 到第 10 站有人坐（利用率是 10−0=10），第二个座位从第 1 到第 9 站有人坐，座位利用率为 (10−0)+(9−1)=18。乘客在某站下车后，其他乘客从这一站就可以开始使用这个座位。保证任意时刻列车上乘客数量不超过 m 即可。

## 输入描述

第一行输入 m、n 和 x 三个数字，分别表示列车座位数量、停靠站点数量和预定乘客数量。

接下来 x 行输入，表示 x 条预定记录，每行有两个输入，分别表示此预定记录的上车站点和下车站点。

## 输出描述

输出一个整数，表示最大的座位利用率。

## 样例 1

> 输入：
> 
> 2 11 4
> 
> 0 1
> 
> 1 9
> 
> 0 10
> 
> 3 8
> 
> 输出：
> 
> 19
> 
> 解释：选择前三位乘客可以使座位利用率最大：19 = (1−0) + (9−1) + (10−0)。若选择后两位乘客，则利用率为 15 = (10−0) + (8−3)。若选择全部四位乘客，则第 3 到第 8 站车上存在 3 名乘客，超出列车座位数。

## 样例 2

> 输入：
> 
> 1 11 2
> 
> 0 5
> 
> 5 10
> 
> 输出：
> 
> 10
> 
> 解释：选择两位乘客可以使座位利用率最大：10 = (5−0) + (10−5)。

## 提示

- $1 <= m <= 9$
- $2 <= n <= 20$
- $1 <= x <= 9$

# 公网下线方案

公有云的某个 region 内，N 个网络节点组网情况可以使用一个 N×N 的矩阵 matrix 表示。在这个组网图中，matrix[i][j] = p 时，表示用户在编号为 i 的节点访问编号为 j 的节点时，必须在 i 节点上具有 ≥ p 的权限等级（p=0 时表示无法通过 i 节点访问 j 节点）。如果用户成功访问了节点，那么它在 j 节点上的权限等级调整为 p。

exposed 为一个整数数组，表示暴露在公网上的网络节点的编号列表。某天扫描发现这批暴露在公网的节点存在被外部恶意攻击风险，且该攻击会影响到可访问的其他节点，并可以持续传递进行攻击。被恶意攻击的节点从公网访问时，攻击者获得了 ROOT 权限（权限等级为 10，即最大值）。

小明是一名网络安全工程师，为了在有限的时间内尽可能减少故障带来的损失，需要立即将某个节点从公网“下线”。假设攻击结束时，被攻击过的节点数量为 R，请帮小明计算出将哪个节点下线能使 R 尽可能小。如果有多个节点满足条件，返回索引最小的那个节点。注意：从公网“下线”的节点不会受到来自公网的攻击，但仍然可能被“可访问”的其他节点传递攻击。

## 输入描述

输入的第一行是网络节点数量 N。

后续的 N 行，每行 N 个数字 v，以空格分割，形成一个 N×N 的矩阵，表示网络节点组网的矩阵。

最后一行，输入 exposed 数组，表示暴露在公网上的网络节点的编号列表。数组元素不会重复。

- 2 ≤ N ≤ 24
- 0 ≤ v ≤ 10
- 0 ≤ exposed[i] ≤ N−1

## 输出描述

输出在 exposed 数组中，计划“下线”的那个节点的编号。

## 样例 1

> 输入：
> 
> 4
> 
> 1 0 0 0
> 
> 0 1 2 0
> 
> 0 1 1 4
> 
> 0 0 3 1
> 
> 1 3
> 
> 输出：
> 
> 3
> 
> 解释：
> 1 和 3 是公网暴露的节点。假设从 1 节点登录：
> - 可以访问 2 节点，权限等级为 2。
> - 从 2 节点可以访问 3 节点，权限等级为 4（因为 matrix[2][3] = 4，当前权限等级为 2，无法访问）。
> 如果将 1 号节点从公网下线，那么攻击者只能通过 3 号节点登录，此时 R 的值为 3。
> 如果将 3 号节点从公网下线，则攻击者只能通过 1 号节点访问到 2 号节点，而 2 号节点无法再访问 3 号节点，此时 R 的值为 2。因此选择 R 值更小的方案，答案为 3。
